As data is very much imbalance so we use hybrid sampling under sample the class 0 to 10000 and over sample the class 1 to class 0

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of normal Transactions --> 31677


```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from collections import Counter

# Define features (X) and target (y)
X = credit_card_data.drop('Class', axis=1) # Replace with your actual target column name
y = credit_card_data["Class"]


# Scale the Features FIRST
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


minority_count = sum(y == 1)

# Step 1: Under-Sample the majority class (Reduce Class 0 to 10,000 samples)
under_sampler = RandomUnderSampler(sampling_strategy={0: minority_count * 10}, random_state=42)
X_under, y_under = under_sampler.fit_resample(X, y)

# Step 2: Over-Sample the minority class (Balance it with Class 0)
smote = SMOTE(sampling_strategy=1.0, random_state=42)  # Fully balance classes
X_balanced, y_balanced = smote.fit_resample(X_under, y_under)

# Check class distribution before & after balancing
print("Original Class Distribution:", Counter(y))  # Example: {0: 31000, 1: 102}
print("Balanced Class Distribution:", Counter(y_balanced))  # Example: {0: 10000, 1: 10000}
```

```python
new_dataset = pd.concat([X_balanced, y_balanced], axis=1)
```

```python
# Splitting the data into Features & Targets
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

from sklearn.preprocessing import StandardScaler
# Initialize the scaler
scaler = StandardScaler()

# Fit on training data and transform both train & test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression(max_iter=1000, solver='saga')
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python


from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


# Define the parameter grids for each model
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}


# Create model instances
models = {
    'SVM': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}


# Iterate through models and perform grid search
for model_name, model in models.items():
    print(f"\nPerforming Grid Search for {model_name}:")

    if model_name == 'SVM':
        grid_search = GridSearchCV(model, param_grid_svm, cv=5, scoring='accuracy')
    elif model_name == 'Decision Tree':
        grid_search = GridSearchCV(model, param_grid_dt, cv=5, scoring='accuracy')
    elif model_name == 'Random Forest':
        grid_search = GridSearchCV(model, param_grid_rf, cv=5, scoring='accuracy')

    grid_search.fit(X_train_scaled, Y_train)

    print("Best Parameters:", grid_search.best_params_)
    print("Best Accuracy:", grid_search.best_score_)

    # Evaluate on the test set using the best model
    best_model = grid_search.best_estimator_
    test_accuracy = best_model.score(X_test_scaled, Y_test)
    print("Test Accuracy:", test_accuracy)
```

```python


# Based on the provided code and the grid search results, the Random Forest model appears to be the best choice.
# Here's the code to train and evaluate the best model (Random Forest with optimized hyperparameters):
# Use the best parameters found from the grid search (adjust accordingly based on your results)
best_rf_params = {'max_depth': None, 'min_samples_split': 2, 'n_estimators': 200}

# Create and train the model with the best parameters
best_model = RandomForestClassifier(**best_rf_params)
best_model.fit(X_train_scaled, Y_train)

# Evaluate the model on the test set
test_accuracy = best_model.score(X_test_scaled, Y_test)
print("Test Accuracy (Best Random Forest):", test_accuracy)
```
